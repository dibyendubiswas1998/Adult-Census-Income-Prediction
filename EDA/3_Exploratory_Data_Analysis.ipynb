# Exploratory Data Analysis (EDA)

```python
import numpy as np
import pandas as pd
import math
import scipy.stats as stats
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
import pylab
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import NearMiss
```

```python
data = pd.read_csv("../DATA/adult_new.csv", sep=',')
data.head()
```

```python
# sex feature:

plt.figure(figsize=(12,6))
sns.countplot(data=data, x="sex", hue='salary')
plt.title("Frequency Count vs Sex based on salary")
plt.xlabel("sex category")
plt.ylabel("frequency")
plt.show()
```

```python
# country feature:

vals = data['country'].value_counts(sort='ascending')[0:10].values
ind = data['country'].value_counts(sort='ascending')[0:10].index

plt.figure(figsize=(102,10))
sns.countplot(data=data, x='country', hue='salary')
plt.title("Frequency Count vs Country feature based on salary")
plt.xlabel("countries")
plt.ylabel("frequency")
plt.show()
```

```python
# workclass feature:

plt.figure(figsize=(12,6))
sns.countplot(data=data, x="workclass", hue='salary')
plt.title("Frequency Count vs Workclass feature based on salary")
plt.xlabel("workclass category")
plt.ylabel("frequency")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=data, x="workclass", hue='sex')
plt.title("Frequency Count vs Workclass feature based on male & female")
plt.xlabel("workclass category")
plt.ylabel("frequency")
plt.show()
```

```python
data['workclass'].value_counts()
```

```python
# marital-status feature:

plt.figure(figsize=(12,6))
sns.countplot(data=data, x="marital-status", hue='sex')
plt.title("Frequency Count vs Marital-Status feature based on male & female")
plt.xlabel("marital-status category")
plt.ylabel("frequency")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=data, x="marital-status", hue='salary')
plt.title("Frequency Count vs Marital-Status feature based on salary")
plt.xlabel("marital-status category")
plt.ylabel("frequency")
plt.show()
```

```python
# education feature

plt.figure(figsize=(12,6))
sns.countplot(data=data, x="education", hue='sex')
plt.title("Frequency Count vs Education feature based on male & female")
plt.xlabel("education category")
plt.ylabel("frequency")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=data, x="education", hue='salary')
plt.title("Frequency Count vs Education feature based on salary")
plt.xlabel("education category")
plt.ylabel("frequency")
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.lineplot(data=data, x="age", y="capital-gain", hue="sex")
plt.title("Capital Gain Vs Age based on male & female")
plt.xlabel("Age")
plt.ylabel("capital gain")
plt.show()
```

```python
plt.figure(figsize=(16,8))
sns.lineplot(data=data, x="age", y="capital-loss", hue="sex")
plt.title("Capital Loss Vs Age based on male & female")
plt.xlabel("Age")
plt.ylabel("capital loss")
plt.show()
```
